```python
import pandas as pd
import test_model_functions

bend_df = pd.read_csv("bend.csv")
```

```python
X = bend_df.Value
Y = bend_df.State
```

```python
c = []
for state in Y:
    if state == 'F':
        c.append('red')
    elif state == 'M':
        c.append('blue')
    else:
        c.append('lightblue')
```

```python
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

classes = ['F','M','N']
class_colours = ['red','blue','lightblue']
recs = []
for i in range(0,len(class_colours)):
    recs.append(mpatches.Rectangle((0,0),1,1,fc=class_colours[i]))
ax1 = plt.axes(frameon=True)
ax1.axes.get_yaxis().set_visible(False)
plt.legend(recs,classes,loc=4)
plt.scatter(X,[1]*len(X), color = c, alpha=0.05, s=1000)
```

```python
N = bend_df['Value'][:100]
M = bend_df['Value'][100:200]
F = bend_df['Value'][200:]
```

```python
print('Min of Neutral: ' + str(min(N)) + 
      '\nMax of Neutral: ' + str(max(N)))
```

```python
print('Min of Second joint: ' + str(min(M)) + 
      '\nMax of Second joint: ' + str(max(M)))
```

```python
print('Min of Both joints: ' + str(min(F)) + 
      '\nMax of Both joints: ' + str(max(F)))
```

```python
print('Area of overlap is (' + str(min(F)) + ", " + str(max(M)) + ")")
```

```python
Overlap_M = [m for m in M if(m >= 62)]
Overlap_F = [f for f in F if(f <= 105)]
```

```python
state_color = ['blue']*(len(Overlap_M)) + ['red']*(len(Overlap_F))
```

```python
ax1 = plt.axes(frameon=True)
ax1.axes.get_yaxis().set_visible(False)
plt.scatter(Overlap_M + Overlap_F, 
            [1]*len(Overlap_M + Overlap_F), 
            color = state_color,
            alpha=0.05, 
            s=2000)

classes = ['F','M']
class_colours = ['red','blue']
recs = []
for i in range(0,len(class_colours)):
    recs.append(mpatches.Rectangle((0,0),1,1,fc=class_colours[i]))
plt.legend(recs,classes,loc=4)

plt.xlabel('Value')
plt.title('Overlapping Region')
```
